# 07 - Final Forecasts (Retrain on Full Data & Predict Future)

Goal:
- Retrain best models using full historical data
- Produce future forecasts (daily & weekly)
- Save forecasts and plots for reporting / deployment


```python
# Cell 1 - imports & paths
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")

from prophet import Prophet

# Directories
DATA_DIR = "../data/processed"
OUT_DIR = "../outputs"
FC_DIR = os.path.join(OUT_DIR, "forecasts")
FIG_DIR = os.path.join(OUT_DIR, "figures")
os.makedirs(FC_DIR, exist_ok=True)
os.makedirs(FIG_DIR, exist_ok=True)

print("Output dirs:", FC_DIR, FIG_DIR)
```

```python
# Cell 2 - configuration
# You can change these before running:
DAILY_HORIZON_DAYS = 90    # forecast next X days
WEEKLY_HORIZON_WEEKS = 26  # forecast next X weeks
INCLUDE_HOLIDAYS = False   # set True to add holidays (see note)
SEASONALITY_MODE = 'multiplicative'  # 'additive' or 'multiplicative'

print("Daily horizon:", DAILY_HORIZON_DAYS, "days")
print("Weekly horizon:", WEEKLY_HORIZON_WEEKS, "weeks")
print("Seasonality mode:", SEASONALITY_MODE)
```

```python
# Cell 3 - load / prepare holidays (optional)
# If you want to include a holidays calendar, prepare a dataframe with columns ['ds','holiday']
# Example (uncomment & adapt for your market):
#
# holidays = pd.DataFrame({
#     'ds': pd.to_datetime(['2017-10-19','2017-12-25','2018-11-07']), # sample dates (yyyy-mm-dd)
#     'holiday': ['diwali','christmas','diwali_2018']
# })
#
# If you don't have holidays, keep holidays=None

holidays = None

if INCLUDE_HOLIDAYS:
    if holidays is None:
        raise ValueError("INCLUDE_HOLIDAYS=True but holidays dataframe is None. Provide holidays or set INCLUDE_HOLIDAYS=False.")
    else:
        print("Using holidays with", len(holidays), "rows")
else:
    print("Holidays disabled (set INCLUDE_HOLIDAYS=True and provide holidays dataframe if needed).")
```

```python
# Cell 4 - Daily Prophet retrain on full data and forecast

# Load full daily series
daily_path = os.path.join(DATA_DIR, "sales_daily.csv")
df_daily = pd.read_csv(daily_path, parse_dates=['order_date'])
df_daily = df_daily.rename(columns={'order_date':'ds','sales':'y'})
df_daily = df_daily.set_index('ds').asfreq('D').reset_index()  # ensure daily freq

print("Daily rows (full):", len(df_daily))
display(df_daily.head())

# Build & fit model on full data
m_daily = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False,
                  seasonality_mode=SEASONALITY_MODE, holidays=holidays)
m_daily.fit(df_daily)
print("Daily Prophet fitted on full data.")

# Forecast future
future_daily = m_daily.make_future_dataframe(periods=DAILY_HORIZON_DAYS, freq='D')
forecast_daily = m_daily.predict(future_daily)

# Save forecast (complete future + history)
out_daily_csv = os.path.join(FC_DIR, f"final_prophet_daily_forecast_{DAILY_HORIZON_DAYS}d.csv")
forecast_daily[['ds','yhat','yhat_lower','yhat_upper']].to_csv(out_daily_csv, index=False)
print("Saved daily forecast to:", out_daily_csv)

# Plot and save
plt.figure(figsize=(12,5))
m_daily.plot(forecast_daily)
plt.title(f"Final Prophet Daily Forecast (next {DAILY_HORIZON_DAYS} days)")
plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR, f"final_prophet_daily_{DAILY_HORIZON_DAYS}d.png"))
plt.show()

# Also save only the future horizon rows (optional)
horizon_start = df_daily['ds'].max() + pd.Timedelta(days=1)
future_horizon_df = forecast_daily[forecast_daily['ds'] >= horizon_start][['ds','yhat','yhat_lower','yhat_upper']]
out_daily_future_csv = os.path.join(FC_DIR, f"final_prophet_daily_horizon_{DAILY_HORIZON_DAYS}d.csv")
future_horizon_df.to_csv(out_daily_future_csv, index=False)
print("Saved daily future-horizon forecast to:", out_daily_future_csv)
```

```python
# Cell 5 - Weekly Prophet retrain on full data and forecast

# Load full weekly series
weekly_path = os.path.join(DATA_DIR, "sales_weekly.csv")
df_weekly = pd.read_csv(weekly_path, parse_dates=['order_date'])
df_weekly = df_weekly.rename(columns={'order_date':'ds','sales':'y'})
df_weekly = df_weekly.set_index('ds').asfreq('W').reset_index()

print("Weekly rows (full):", len(df_weekly))
display(df_weekly.head())

# Build & fit model on full data
m_weekly = Prophet(yearly_seasonality=True, weekly_seasonality=False,
                   seasonality_mode=SEASONALITY_MODE, holidays=holidays)
m_weekly.add_seasonality(name='yearly_custom', period=52, fourier_order=5)
m_weekly.fit(df_weekly)
print("Weekly Prophet fitted on full data.")

# Forecast future
future_weekly = m_weekly.make_future_dataframe(periods=WEEKLY_HORIZON_WEEKS, freq='W')
forecast_weekly = m_weekly.predict(future_weekly)

out_weekly_csv = os.path.join(FC_DIR, f"final_prophet_weekly_forecast_{WEEKLY_HORIZON_WEEKS}w.csv")
forecast_weekly[['ds','yhat','yhat_lower','yhat_upper']].to_csv(out_weekly_csv, index=False)
print("Saved weekly forecast to:", out_weekly_csv)

# Plot and save
plt.figure(figsize=(12,5))
m_weekly.plot(forecast_weekly)
plt.title(f"Final Prophet Weekly Forecast (next {WEEKLY_HORIZON_WEEKS} weeks)")
plt.tight_layout()
plt.savefig(os.path.join(FIG_DIR, f"final_prophet_weekly_{WEEKLY_HORIZON_WEEKS}w.png"))
plt.show()

# Save only future horizon rows
horizon_start_w = df_weekly['ds'].max() + pd.Timedelta(weeks=1)
future_weekly_df = forecast_weekly[forecast_weekly['ds'] >= horizon_start_w][['ds','yhat','yhat_lower','yhat_upper']]
out_weekly_future_csv = os.path.join(FC_DIR, f"final_prophet_weekly_horizon_{WEEKLY_HORIZON_WEEKS}w.csv")
future_weekly_df.to_csv(out_weekly_future_csv, index=False)
print("Saved weekly future-horizon forecast to:", out_weekly_future_csv)
```
